```python
# ==========================================
# TASK 01: DATA VISUALIZATION
# ==========================================
# Objective:
# Create a histogram to visualize the distribution
# of population across countries using the
# World Bank population dataset.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load World Bank population dataset
url = "https://raw.githubusercontent.com/Prodigy-InfoTech/data-science-datasets/main/Task%201/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"

df = pd.read_csv(url, skiprows=4)

# Preview dataset
df.head()
```

```python
# Check dataset information
df.info()
```

```python
# Select population data for year 2022
population_2022 = df[['Country Name', '2022']].dropna()

# Rename column for clarity
population_2022.columns = ['Country', 'Population']

population_2022.head()
```

```python
# Summary statistics
population_2022['Population'].describe()
```

```python
plt.figure(figsize=(8, 5))

plt.hist(population_2022['Population'], bins=20)

plt.title("Distribution of Population Across Countries (2022)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")

plt.show()
```

```python
print("Insights:")
print("1. The population distribution is highly right-skewed.")
print("2. Most countries have relatively small populations.")
print("3. Only a few countries have very large populations, acting as outliers.")
```
